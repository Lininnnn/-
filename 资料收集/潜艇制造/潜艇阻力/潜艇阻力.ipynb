### 1. **总阻力公式**

总阻力 $R_t$ 由摩擦阻力 $R_f$ 和波浪阻力 $R_w$ 组成：

$$
R_t = R_f + R_w
$$

- $R_f$ 是摩擦阻力，主要与潜艇表面与流体的相互作用有关。
- $R_w$ 是波浪阻力，主要由潜艇在密度分层流体中激发的自由表面波和内部波引起。


### 2. **摩擦阻力公式**

摩擦阻力是通过经验公式或使用CFD模型计算的。由于文中使用的是基于湍流模型（Realizable k-ε）的CFD方法，没有直接给出具体的经验公式。

### 3. **波浪阻力公式**

波浪阻力 $R_w$ 使用平面波理论计算。对于潜艇在自由表面或内部表面生成的波，波浪阻力 $R_w$ 表达为：

$$
R_w = \frac{1}{16} \rho g H^2 b
$$

其中：

- $\rho$ 是流体密度，
- $g$ 是重力加速度，
- $H$ 是波高，
- $b$ 是波宽（波能量区域的宽度）。

波浪阻力系数 $C_w$ 可表示为：

$$
C_w = \frac{R_w}{\frac{1}{2} \rho v^2 S}
$$

其中：

- $v$ 是潜艇的前进速度，
- $S$ 是湿表面积，
- $L$ 是潜艇的长度。


### 4. **波浪阻力系数公式**

根据平面波理论，波浪阻力系数 $C_w$ 的近似公式为：

$$
C_w = \left[ C + D \cos \left( \frac{2 \pi m L}{\lambda} \right) \right] \left( \frac{v}{\sqrt{gL}} \right)^4
$$

其中：

- $\lambda$ 是波长，
- $m$ 是与船型和弗劳德数 $Fr$ 相关的系数，
- $C$ 和 $D$ 是常数。


### 5. **弗劳德数公式**

弗劳德数 $Fr$ 用于描述潜艇的速度与波浪特性之间的关系：

$$
Fr = \frac{v}{\sqrt{gL}}
$$

其中：

- $v$ 是潜艇的前进速度，
- $L$ 是潜艇的长度，
- $g$ 是重力加速度。

这些公式用于描述潜艇在密度分层流体中的水动力特性，尤其是速度、潜深、密度等参数对阻力的影响。
