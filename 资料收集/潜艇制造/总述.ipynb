### 材料选择
建造潜艇将使用哪些材料？

### 推进系统
将使用什么类型的推进系统？  
它会是柴电系统、核反应堆还是螺旋桨驱动系统？

### 设计完成后，施工阶段开始。此阶段涉及建造潜艇的船体、推进系统和其他组件。施工过程通常包括几个阶段，包括：

### 船体结构
潜艇的船体采用钢、铝和复合材料的组合建造。  
船体的设计可以承受深水的压力。

### 推进系统安装
推进系统，包括发动机、发电机和螺旋桨，安装在潜艇内。

### 控制系统安装
控制系统，包括方向舵、转向和导航系统，都安装在潜艇中。

### 电气系统安装
电气系统，包括电池、发电机和配电板，都安装在潜艇中。

### 关键组件几个关键部件对潜艇的运行至关重要：

### 压载舱
压载舱用于控制潜艇的浮力，并允许其俯冲和浮出水面。

### 推进轴
推进轴用于将动力从发动机传递到螺旋桨。

### 控制面
控制面（包括方向舵和平面）用于控制潜艇的方向和深度。

### 传感器和导航系统
传感器和导航系统，包括声纳和 GPS，用于检测和导航潜艇的周围环境。
